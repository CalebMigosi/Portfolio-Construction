```python
def get_ind_filetype(filetype):
    """
    Load and format the ken french 30 industry portfolios profiles types
    """
    types = ["returns", "nfirms", "size"]
    if filetype not in types:
        sep = ", "
        raise ValueError(f"The filetype should be one of {sep.join(types)}")

    if filetype == "returns":
        filename = "vw_rets"
        divisor = 100
    elif filetype == "nfirms":
        filename = "nfirms"
        divisor = 1
    elif filetype == "size":
        filename = "size"
        divisor = 1

    ind = pd.read_csv(f"ind30_m_{filename}.csv", header=0, parse_dates=True,
                      index_col=0) / divisor
    ind.index = pd.to_datetime(ind.index, format='%Y%m')
    ind.index = ind.index.to_period('M')
    ind.columns = ind.columns.str.strip()

    return ind
```

```python
#Get portfolio returns 
returns_data = get_ind_filetype("returns")
sector_list = ["Food", "Beer", "Smoke", "Games", "Fin"]
```

```python
returns_data = returns_data.loc["1990":"2018", sector_list]
portfolio = portfolio_statistics(returns_data)
```
